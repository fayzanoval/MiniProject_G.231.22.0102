```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
```

```python
data = pd.read_csv("stock.csv")
```

```python
data
```

```python
sns.histplot(data, x="Purchase Price", color='c', edgecolor='white')
plt.title("Histogram of Purchase Price", fontstyle='oblique', fontsize=25, color='c', weight='semibold', family='monospace')
plt.xlabel("Purchase Price", fontsize=20, color='c', weight='semibold', family='monospace')
plt.ylabel("Count", fontsize=20, color='c', weight='semibold', family='monospace')
plt.show()

sns.histplot(data, x="Current Price", color='c', edgecolor='white')
plt.title("Histogram of Current Price", fontstyle='oblique', fontsize=25, color='c', weight='semibold', family='monospace')
plt.xlabel("Current Price", fontsize=20, color='c', weight='semibold', family='monospace')
plt.ylabel("Count", fontsize=20, color='c', weight='semibold', family='monospace')
plt.show()

sns.histplot(data, x="Gain/Loss", color='c', edgecolor='white')
plt.title("Histogram of Gain/Loss", fontstyle='oblique', fontsize=25, color='c', weight='semibold', family='monospace')
plt.xlabel("Gain/Loss", fontsize=20, color='c', weight='semibold', family='monospace')
plt.ylabel("Count", fontsize=20, color='c', weight='semibold', family='monospace')
plt.show()
```

```python
top_dividend = data.nlargest(5, 'Dividend/share')[['Stock Name', 'Symbol', 'Dividend/share']]
print('Top 5 Stocks with highest Dividend/share:')
print(top_dividend)
```

```python
top_market_value = data.nlargest(5, 'Market Value')[['Stock Name', 'Symbol', 'Market Value']]
print('\nTop 5 Stocks with highest Market Value:')
print(top_market_value)
```

```python
top_annual_yield = data.nlargest(7, 'Annual Yield')[['Stock Name', 'Symbol', 'Annual Yield']]
print('\nTop 7 Stocks with highest Annual Yield:')
print(top_annual_yield)
```
